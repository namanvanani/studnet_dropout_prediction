```python
import pandas as pd
import numpy as np
```

```python
with open("new_text.txt", "r") as f:
    ans = f.readlines()
```

```python
ans = [i.split('\t') for i in ans]
```

```python
df = pd.DataFrame()
```

```python
df[['State', 'Primary', 'Upper Primary', 'Secondary']] = [[i[0], i[3], i[6], i[9][:-1]] for i in ans]
```

```python
df[['State','Secondary']].to_csv('primary.txt', header=None, index=None, sep=' ', mode='a')
```
