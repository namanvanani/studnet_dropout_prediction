```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
generated_data = pd.read_csv("student_dropout_data.csv", index_col='id')
generated_data
```

```python
cols_to_encode = ["sex", "urban/rural", "income", "Parents Qualification", "is dropout?"]
```

```python
df_encoded = pd.get_dummies(generated_data, columns=cols_to_encode, drop_first=True)
```

```python
df_encoded
```

```python
import pickle
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

# df_encoded[["GPA", "average attendance"]]=scaler.fit(df_encoded[["GPA", "average attendance"]])
scaler.fit(df_encoded[['GPA', 'average attendance']])
df_encoded[["GPA", "average attendance"]] = scaler.transform(df_encoded[["GPA", "average attendance"]])

with open("scaler_model.pkl", "wb") as f:
    pickle.dump(scaler, f)
```

```python
with open('scaler_model.pkl', 'rb') as file:
    new_scaler = pickle.load(file)
    
new_scaler.transform([[4,80]])
```

```python
df_encoded
```

# Logistic Regression

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

X = df_encoded.drop(columns=["name", "is dropout?_Yes"])
y = df_encoded['is dropout?_Yes']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)

model = LogisticRegression()
model.fit(X_train, y_train)

predictions = model.predict(X_test)

accuracy = accuracy_score(y_test, predictions)
print("Logistic Regression Accuracy:", accuracy)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, classification_report

cm = confusion_matrix(y_test, predictions)
cm
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Predicted 0', 'Predicted 1'],
            yticklabels=['Actual 0', 'Actual 1'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.show()
```

```python
print(classification_report(y_test, predictions))
```

# After Equalizing the Values

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(sampling_strategy='auto', random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, stratify=y_resampled)
```

# Logistic Regression Model

```python
model = LogisticRegression()
model.fit(X_train, y_train)

predictions = model.predict(X_test)
    
accuracy = accuracy_score(y_test, predictions)
print("Logistic Regression Accuracy:", accuracy)
```
