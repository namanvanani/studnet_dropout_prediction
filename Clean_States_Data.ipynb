```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("DOR.csv")
def clean_state(x):
    if len(x.split(" ")) == 3:
        ls = x.split(" ")
        return " ".join([ls[0], ls[-1]])
    elif x=="All India":
        return "India"
    else:
        return x
df["State_UT"] = df["State_UT"].apply(clean_state)
cols_to_int = ['Primary_Boys', 'Primary_Girls', 'Primary_Total',
       'Upper Primary_Boys', 'Upper Primary_Girls', 'Upper Primary_Total',
       'Secondary _Boys', 'Secondary _Girls', 'Secondary _Total',
       'HrSecondary_Boys', 'HrSecondary_Girls', 'HrSecondary_Total']
for column in cols_to_int:
    df[column] = pd.to_numeric(df[column], errors='coerce')

new_cols = ["State", "Year", "PB", "PG", "PT", "UPB", "UPG", "UPT", "SB", "SG", "ST", "USB", "USG", "UST"]
df = df.rename(columns=dict(zip(df.columns, new_cols)))
df.Year = df.Year.apply(lambda x:int(x.split('-')[-1])+2000)
```

```python
df_filled = pd.DataFrame(df.iloc[:,2:]).apply(lambda row: row.fillna(row.mean()), axis=1)
```

```python
new_df = pd.concat([df[["State", "Year"]], df_filled], axis=1)
```

```python
new_df.to_csv("final_state_data_file.csv")
```
